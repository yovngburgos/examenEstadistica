2. Calcula e interpreta el coeficiente de asimetr칤a y el coeficiente de curtosis para la variable Pago mensual. 쮼stos valores son consistentes con los esperados en una distribuci칩n normal? Justifica tu respuesta.

```python
from scipy.stats import skew, kurtosis
asimetria = skew(base_datos['Pago mensual ($)'])
curtosis = kurtosis(base_datos['Pago mensual ($)'])
print(f'El coeficiente de asimetr칤a es: {asimetria}')
print(f'El coeficiente de curtosis es: {curtosis}')
```

**Respuesta:** Como la curtosis est치 entre -0,5 y 0,5 (variable mesoc칰rtica) y el coef. de asimetr칤a est치 entre estos mismos valores (variable sim칠trica) podemos decir que la variable pago mensual puede ser de distribuci칩n normal

4. Considerando que la variable Pago mensual sigue una distribuci칩n aproximadamente normal, responde las siguientes preguntas:

$\hspace{1cm}$ a)  쮺u치l es la probabilidad de que un cliente pague menos de \$18.000 al mes?

$\hspace{1cm}$ b) 쮺u치l es la probabilidad de que un cliente pague m치s de \$25.000 al mes?

$\hspace{1cm}$ c) 쮺u치l es la probabilidad de que un cliente pague entre \$20.000 y \$24.000 al mes?

```python
from scipy.stats import norm

media= base_datos['Pago mensual ($)'].mean()
desviacion= base_datos['Pago mensual ($)'].std()

#a)
#P(X<18000)
ra = norm.cdf(18000, media, desviacion)*100

#b)
#P(X>25000)
rb = (1-norm.cdf(25000, media, desviacion))*100

#c)
#P(20000<X<24000)
rc = (norm.cdf(24000, media, desviacion) - norm.cdf(20000, media, desviacion))*100

print(f'a) La probabilidad es de {round(ra,2)}%')
print(f'b) La probabilidad es de {round(rb,2)}%')
print(f'c) La probabilidad es de {round(rc,2)}%')
```
