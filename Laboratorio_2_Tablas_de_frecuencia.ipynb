3. Crea una tabla de frecuencias puntuales y acumuladas utilizando la variable *Puntaje de felicidad*. Considera crear la tabla de frecuencias con 8 intervalos.

```python
import pandas as pd
base_datos= pd.read_csv('Reporte_felicidad_2024.csv')
```

```python
base_datos.describe()
```

```python
base_datos.head(1)
```

```python
#Tama침o de la base de datos
import pandas as pd

tama침o=base_datos.shape[0]

base_datos['Intervalos_variable'] = pd.cut(base_datos['Puntaje de felicidad'], bins = 8, include_lowest= True, precision = 0)

cuenta_intervalos_variable= base_datos.groupby('Intervalos_variable', observed = True).size()

tabla= pd.DataFrame({
    'Frecuencia absoluta (fi)': cuenta_intervalos_variable,
    'Frecuencia absoluta acumulada (Fi)': cuenta_intervalos_variable.cumsum(),
    'Frecuencia relativa (%) (hi)': cuenta_intervalos_variable/tama침o * 100,
    'Frecuencia relativa acumulada (%) (Hi)': (cuenta_intervalos_variable/tama침o * 100).cumsum()
})
round(tabla, 1)
```
